## Cython과 Numba을 사용한 속도 향상(컴파일러로 만듦)
### cython
1. CPython 확장 모듈을 손쉽게 생성하도록 고안된 컴파일 언어
2. 파이썬 문법을 기반으로 C/C++루틴을 호출
3. 파이썬(스크립트언어), C언어(컴파일언어-소스코드를 미리 번역해둔 실행 파일을 수행하기 때문에 실시간으로 코드를 해석할 필요없어 수행 속도가 빠르다.)
4. 도입배경 : 파이썬-동적 프로그래밍 언어라서 매번 변수 타입을 확인해야 하고 반복문과 사용될 경우 시간 손실이 더 많다.
4-1. 이렇게 파이썬이 가진 실시간 코드 해석 특성과 동적 변수 특성 등으로 인한 수행 속도의 한계를 극복하기 위해 '파이썬 스크립트를 C언어로 컴파일'한다.
5. 원리 : C언어 코드를 부르기 위해 변수를 전달하고 결과를 전달 받는 필요한 추가적인 코드 의미함. 
5-1. 싸이썬으로 C언어 함수 호출 가능하다.
6. sudo pip install cython
